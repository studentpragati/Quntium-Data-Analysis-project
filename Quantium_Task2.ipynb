###  Define Trial and Control Stores

```python
# Define trial stores manually
trial_stores= [77,86,88]

# Select data for trial stores
trial_data= df[df["STORE_NBR"].isin(trial_stores)]

# Select the remaining stores as control stores
control_data = df[~df["STORE_NBR"].isin(trial_stores)]
```

### Visualize Sales Comparison by Store Type

#### Group by STORE_NBR to get total sales for each store


```python
# Group the data by store and sum the total sales
total_sales_by_store = df[["STORE_NBR", "TOT_SALES"]].groupby("STORE_NBR").sum().reset_index()

# Filter for trial stores
trial_sales = total_sales_by_store[total_sales_by_store["STORE_NBR"].isin(trial_stores)]

# Filter for control stores
control_sales = total_sales_by_store[~total_sales_by_store["STORE_NBR"].isin(trial_stores)]
```

### Plot the Sales for Trial and Control Stores Separately

```python
# Plot for trial stores
plt.figure(figsize=(8, 5))
sns.barplot(x="STORE_NBR", y="TOT_SALES", data=trial_sales, palette="Blues_d")
plt.title('Total Sales for Trial Stores')
plt.xlabel('Store Number')
plt.ylabel('Total Sales')
plt.show()

# Plot for control stores
plt.figure(figsize=(8, 5))
sns.barplot(x="STORE_NBR", y="TOT_SALES", data=control_sales, palette="Greens_d")
plt.title('Total Sales for Control Stores')
plt.xlabel('Store Number')
plt.ylabel('Total Sales')
plt.show()
```

Explanation:
Grouping: We group the data using df[["STORE_NBR", "TOT_SALES"]].groupby("STORE_NBR").sum() to calculate the total sales for each store.
Separate Plots: We plot the sales for trial stores and control stores separately to avoid clutter and make the plots more readable.
Bar Plot: The sns.barplot() is used to create a simple bar plot with store numbers on the x-axis and total sales on the y-axis.

### Compare Trial Stores to Control Stores
##### Let's compare the total sales of each trial store with the average sales of control stores

```python
results = []

# Calculate the average sales for control stores
control_avg_sales = control_data.groupby('STORE_NBR')['TOT_SALES'].sum().mean()

for trial in trial_stores:
    # Get data for the current trial store
    trial_store_data = df[df['STORE_NBR'] == trial]
    
    # Calculate total sales for the trial store
    trial_total_sales = trial_store_data['TOT_SALES'].sum()
    
    # Store the results for comparison
    results.append({
        'trial_store': trial,
        'trial_total_sales': trial_total_sales,
        'control_avg_sales': control_avg_sales
    })

# Print results for each trial store
for result in results:
    print(f"Trial Store: {result['trial_store']}, Trial Sales: {result['trial_total_sales']}, Control Avg Sales: {result['control_avg_sales']}")
```

###  Summarize Findings

```python
summary = []

for result in results:
    trial_store = result['trial_store']
    trial_sales = result['trial_total_sales']
    control_avg_sales = result['control_avg_sales']
    
    # Check if the trial store sales are higher than control store average sales
    if trial_sales > control_avg_sales:
        success_status = "Successful"
    else:
        success_status = "Not Successful"
    
    # Store the summary
    summary.append({
        'trial_store': trial_store,
        'success_status': success_status,
        'trial_sales': trial_sales,
        'control_avg_sales': control_avg_sales
    })

# Print summary
for item in summary:
    print(f"Trial Store: {item['trial_store']}, Status: {item['success_status']}, "
          f"Trial Sales: {item['trial_sales']}, Control Avg Sales: {item['control_avg_sales']}")
```

### Visualize the Results
#### Visualize the comparison of total sales for each trial store versus the average sales of control stores.

```python
# Convert summary to DataFrame for visualization
visualization_data = pd.DataFrame(summary)

# Create a bar plot to compare trial store sales with control average sales
plt.figure(figsize=(10, 6))
bar_width = 0.35
x = np.arange(len(visualization_data))

# Plot trial store sales and control average sales
plt.bar(x, visualization_data['trial_sales'], width=bar_width, label='Trial Store Sales')
plt.bar(x + bar_width, visualization_data['control_avg_sales'], width=bar_width, label='Control Avg Sales')

plt.xlabel('Trial Stores')
plt.ylabel('Total Sales')
plt.title('Comparison of Trial Store Sales vs Control Store Avg Sales')
plt.xticks(x + bar_width / 2, visualization_data['trial_store'])
plt.legend()
plt.show()
```
